```python
from google.cloud import translate  # Imports the Google Cloud client library
import pandas as pd

print('translate version:', translate.__version__)
print('Pandas version:',pd.__version__)
```

```python
def initialise_parameters():
    global tempTranslateClient
    # Instantiates a client
    tempTranslateClient = translate.Client()

initialise_parameters()  #initialise once

def print_output(translation):
    print(u'Input Text: {}\n'.format(translation['input']))
    print(u'Translated Text: {}\n'.format(translation['translatedText']))
  #  print('Confidence: {}'.format(translation['confidence']))
    print('Detected Source Language: {}'.format(translation['detectedSourceLanguage']))

def translator(txtText, txtTargetLanguage):  
    global tempTranslateClient
    # Translates some text into Russian
    translation = tempTranslateClient.translate(
        txtText,
        target_language=txtTargetLanguage)

    return translation
```

```python
# The text to translate
txtText = u'You can work with either one or reap the benefits of both products in your workflow. Use AutoML Translation for training, managing, and predicting translations from custom models. For all of your other translation needs, leverage Translation API and some of its new sophisticated features, including glossary, batch translation, and model selection.'
# The target language
txtTargetLanguage = 'hi'
translation = translator(txtText, txtTargetLanguage)
print_output(translation)
```

### Code to read text files and line by line translation 

```python
pdHoldOutput = pd.DataFrame(columns=["textLine", "Translation", "DetectedSourceLang", "ConvertedTo"])
def set_path_to_file(tempFileName):
    return './'+tempFileName

def read_convert_lin_by_line(tempPathToTextFile, txtTargetLanguage):
    global pdHoldOutput
    para = open(tempPathToTextFile, 'r')
    i = 0
    for line in para:
        translation = translator(line, txtTargetLanguage)
        print_output(translation)
        
        pdHoldOutput.at[i,'textLine'] = line
        pdHoldOutput.at[i,'Translation'] = translation['translatedText']
        pdHoldOutput.at[i,'DetectedSourceLang'] = translation['detectedSourceLanguage']
        i = i + 1
    pdHoldOutput['ConvertedTo'] = txtTargetLanguage

def translate_text_file(tempFileName,txtTargetLanguage):
    global pdHoldOutput
    
    #set file path
    tempPathToTextFile = set_path_to_file(tempFileName)
    
    #read text file and convert line by line
    read_convert_lin_by_line(tempPathToTextFile, txtTargetLanguage)
    print('Total number of lines converted = ', len(pdHoldOutput))
    return pdHoldOutput
```
